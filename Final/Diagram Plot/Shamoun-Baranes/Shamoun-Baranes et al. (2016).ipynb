```python
#'Shamoun-Baranes et al. (2016).csv'
import joblib

# Loading the Data produced by model
data_time_load = joblib.load('datetime.pkl')
queue_output_test_load = joblib.load('truth_data.txt')
test_predict_load = joblib.load('predict_data.txt')
```

```python
# Plotting the diagram
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
data_time = data_time_load
plt.plot(data_time,queue_output_test_load[:,0],color='red',label='Test set output-longitude')
plt.plot(data_time,test_predict_load[:,0],color='green',label='Predict-longitude')
plt.legend()

plt.plot(data_time,queue_output_test_load[:,1],color='black',label='Test set output-latitude')
plt.plot(data_time,test_predict_load[:,1],color='orange',label='Predict-latitude')
plt.legend()

plt.xticks(rotation=90)
ax.set_xlabel(r'Time')
ax.set_ylabel(r'Degree')

plt.show()
fig.savefig('3.jpg', dpi=300, bbox_inches='tight')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Input the data
data = pd.read_csv('Shamoun-Baranes et al. (2016).csv');

b=data.shape;   #shape

# filter out one animols
animal_BirdID_317 = data[data['BirdID']==317]

import time
import datetime
# timestamp
animal_BirdID_317.DateTime = pd.to_datetime(animal_BirdID_317.DateTime)

# set index (datetime)
animal_BirdID_317 = animal_BirdID_317.set_index('DateTime')
animal_BirdID_317.head()

# sort the index
animal_BirdID_317.sort_index(ascending = True).head()

# the location of animal
animal_BirdID_317_pre_data = animal_BirdID_317[animal_BirdID_317.columns[3:5]]

a = animal_BirdID_317_pre_data.resample('D').mean()
animal_BirdID_317_data = a.resample('D').interpolate('linear')
```

```python
year_2010_longitude_data=[]
year_2010_latitude_data=[]

year_2011_longitude_data=[]
year_2011_latitude_data=[]

year_2012_longitude_data=[]
year_2012_latitude_data=[]

year_2013_longitude_data=[]
year_2013_latitude_data=[]

year_2014_longitude_data=[]
year_2014_latitude_data=[]


for i in range(365):

    year_2010_longitude_data.append(animal_BirdID_317_data['Longitude'][i])
    year_2010_latitude_data.append(animal_BirdID_317_data['Latitude'][i])


    year_2011_longitude_data.append(animal_BirdID_317_data['Longitude'][i+365])
    year_2011_latitude_data.append(animal_BirdID_317_data['Latitude'][i+365])


    year_2012_longitude_data.append(animal_BirdID_317_data['Longitude'][i+365+365])
    year_2012_latitude_data.append(animal_BirdID_317_data['Latitude'][i+365+365])


    year_2013_longitude_data.append(animal_BirdID_317_data['Longitude'][i+365+365+365])
    year_2013_latitude_data.append(animal_BirdID_317_data['Latitude'][i+365+365+365])
    
    year_2014_longitude_data.append(animal_BirdID_317_data['Longitude'][i+365+365+365+365])
    year_2014_latitude_data.append(animal_BirdID_317_data['Latitude'][i+365+365+365+365])
        
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111)

#Define the sise of map
mymap = Basemap(llcrnrlon=-15, llcrnrlat=30, urcrnrlon=10, urcrnrlat=55, resolution='h', projection='merc')

mymap.drawparallels(np.arange(30,55,5),labels=[1, 1, 1, 1])
mymap.drawmeridians(np.arange(-15,10,5),labels=[1, 1, 1, 1])

mymap.drawcoastlines()
mymap.fillcontinents()

# 2010
long_1=year_2010_longitude_data
lat_1=year_2010_latitude_data

xpt_1, ypt_1 = mymap(long_1, lat_1)
mymap.plot(xpt_1, ypt_1, 'c*', markersize=5, c='red',linestyle='-',label='Period 1')

# 2011
long_2=year_2011_longitude_data
lat_2=year_2011_latitude_data

xpt_2, ypt_2 = mymap(long_2, lat_2)
mymap.plot(xpt_2, ypt_2, 'c*', markersize=5, c='blue',linestyle='-',label='Period 2')

# 20012
long_3=year_2012_longitude_data
lat_3=year_2012_latitude_data

xpt_3, ypt_3 = mymap(long_3, lat_3)
mymap.plot(xpt_3, ypt_3, 'c*', markersize=5, c='black',linestyle='-',label='Period 3')

# 2013
long_4=year_2013_longitude_data
lat_4=year_2013_latitude_data

xpt_4, ypt_4 = mymap(long_4, lat_4)
mymap.plot(xpt_4, ypt_4, 'c*', markersize=5, c='orange',linestyle='-',label='Period 4')

# 2014-2015(testing)
long_5=year_2014_longitude_data
lat_5=year_2014_latitude_data

xpt_5, ypt_5 = mymap(long_5, lat_5)
mymap.plot(xpt_5, ypt_5, 'c*', markersize=5, c='yellow',linestyle='-',label='Period 5')

# # Predict
# long_6=test_predict_load[:,0]
# lat_6=test_predict_load[:,1]

# xpt_6, ypt_6 = mymap(long_6, lat_6)
# mymap.plot(xpt_6, ypt_6, 'c*', markersize=5, c='green',linestyle='-',label='Predict')

# # testing
# long_7=queue_output_test_load[:,0]
# lat_7=queue_output_test_load[:,1]

# xpt_7, ypt_7 = mymap(long_7, lat_7)
# mymap.plot(xpt_7, ypt_7, 'c*', markersize=5, c='purple',linestyle='-',label='Test')

plt.legend(loc='upper left')
plt.show()
fig.savefig('4.jpg', dpi=300, bbox_inches='tight')
```
